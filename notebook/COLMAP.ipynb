# Ejecución de pruebas

## cereal

```python
!mkdir /content/results/cereal
!colmap \
  feature_extractor \
  --database_path /content/results/cereal/database.db \
  --image_path /content/dataset/cereal/images
!colmap \
  exhaustive_matcher \
  --database_path /content/results/cereal/database.db
!mkdir /content/results/cereal/sparse
!colmap \
  mapper \
  --database_path /content/results/cereal/database.db \
  --image_path /content/dataset/cereal/images \
  --output_path /content/results/cereal/sparse/

!mkdir /content/results/cereal/dense

!colmap image_undistorter \
  --image_path /content/dataset/cereal/images \
    --input_path "/content/results/cereal/sparse/0" \
    --output_path /content/results/cereal/dense \
    --output_type COLMAP \
    --max_image_size 2000

!colmap patch_match_stereo \
    --workspace_path /content/results/cereal/dense \
    --workspace_format COLMAP \
    --PatchMatchStereo.geom_consistency true

!colmap stereo_fusion \
    --workspace_path /content/results/cereal/dense \
    --workspace_format COLMAP \
    --input_type geometric \
    --output_path /content/results/cereal/dense/fused.ply


```
